```python
import numpy as np
import matplotlib.pyplot as plt

def binary_cross_entropy(y_true, y_pred):
    return -(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Các giá trị thực tế (y_true)
y_trues = [1]

# Tạo các giá trị dự đoán (y_pred) từ 0 đến 1
y_preds = np.linspace(0.01, 0.99, 100)

plt.figure(figsize=(10, 6))

for y_true in y_trues:
    losses = binary_cross_entropy(y_true, y_preds)
    plt.plot(y_preds, losses, label=f'y_true = {y_true}')

plt.title('Binary Cross-Entropy Loss')
plt.xlabel('Predicted Probability')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
