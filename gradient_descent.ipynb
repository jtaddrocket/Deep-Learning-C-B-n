Gradient descent là thuật toán tìm giá trị nhỏ nhất của hàm số $f(x)$ dựa trên đạo hàm. Thuật toán:
- Khởi tạo giá trị x = $x_0$ tuỳ ý. 
- Gán x = x - learning_rate*$f'(x)$ (learning_rate > 0).
- Tính lại $f(x)$. Nếu $f(x)$ đủ nhỏ thì dừng lại, ngược lại tiếp tục bước 2.

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = 1000
y = []

for i in range(100):
    x = x - 0.01 * 2 * x #learning_rate = 0.1
    y.append(x**2)

plt.plot(y)
plt.xlabel('Số lần')
plt.ylabel('f(x)')
plt.title('Giá trị f(x) sau số lần thực hiện bước 2')
plt.show()
```

Việc chọn learning_rate cực kỳ quan trọng: 
- Nếu learning_rate nhỏ: mỗi lần hàm số giảm rất ít nên cần nhiều lần thực hiện bước 2 để hàm số đạt giá trị nhỏ nhất. 
- Nếu learning_rate hợp lý: sau một số lần lặp lại bước 2 vừa phải thì hàm sẽ đạt giá trị đủ nhỏ. 
- Nếu learning_rate quá lớn: sẽ gây ra hiện tượng overshoot và sẽ không bao giờ đạt được giá trị nhỏ nhất
